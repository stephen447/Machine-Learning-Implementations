```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from sklearn.linear_model import LinearRegression
import mlUtils as ml
```

```python
df = pd.read_csv("dataTestwithout0.csv")
```

```python
df.head()
```

```python
# inputs
features = df[["make", "model", "year", "mileage", "fuelType", "engineSize"]]
# output/target
y = df["price"]
```

```python
# Normalise the input
x = df["mileage"]
x_norm = (x - x.mean()) / x.std()
```

```python
plt.scatter(x, y, s=10) 
plt.title("Mileage vs Price")
plt.xlabel("Mileage") 
plt.ylabel("Price")          
plt.show()
```

```python
w_final, b_final, J_hist, p_hist = ml.train(x_norm ,y)
```

```python
plt.plot(J_hist)
plt.xlabel("Iteration")
plt.ylabel("Cost J")
plt.title("Cost vs Iteration")
plt.show()
```

```python
# Denormalise the input and parameters
x_mean = x.mean()
x_std = x.std()
w = w_final / x_std
b = b_final - (w_final * x_mean / x_std)

print("Final W:", w)
print("Final B:", b)
```

```python
plt.scatter(x, y, s=10)  # your data points

# predicted line
y_pred = w * x + b
plt.plot(x, y_pred)

plt.xlabel("Mileage")
plt.ylabel("Target")
plt.title("Linear Regression Fit")
plt.show()
```

```python
# Validate that manual coefficients caluclation was correct
# reshape to (m, 1)
X = x.values.reshape(-1, 1)
Y = y.values

model = LinearRegression()
model.fit(X, Y)

w = model.coef_[0]
b = model.intercept_
print("SkitLearn coefficients", w, b)
```
